# Indexing is an important concept in python

```python
print(A[0]) # look at first row ... indexing starts at 0
print(A[1]) # look at second row
print(A[1,:]) # look at second row
print(A[1,0]) # look at element in second row of first column
print(A[:,0]) # look at first column
print(A[:,-1]) # look at last column
print(A[:,:2]) # look at first two columns
```

# Slicing is also an important concept


```python
# Slicing numpy arrays creates a pointer to the original data
# this is different from python lists
A = np.array([[1, 2, 3], [4, 5, 6]])
print(A)
b = A[:,0:2] # create a slice of A (pointing to all rows, first two cols)
print(b)
b[0,0] = 5
print(b)
print(A)
c = A[:].copy() # do this if you need a new, independent copy!
c[1,0] =-1
print(c)
print(A)
```

# Let's review some linear algebra

```python
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[7, 8, 9], [-1, -2, -3]])
# a simple operation: A+B
print(A+B)
# We can make a prettier printing of variables 
print(" A:\n{}\n B:\n{}\n A+B:\n{}".format(A,B,A+B))
```

```python
A.dot(B)
```

## Learn to interpret errors!

```python
print(" A:\n{}\n B:\n{}\n".format(A,B))
print("A.B^T:\n{}\n".format(A.dot(B.T))) # dot product (inner products of rows)
print("A^T.B:\n{}\n".format((A.T).dot(B))) # another dot product (inner products of columns)
print("A circ. B:\n{}\n".format(np.multiply(A,B))) # element-wise multiply
```

# Matplotlib

- Just looking at the numbers in the arrays may often not tell us too much! 
- We need a library to plot and visualize data, which is the matplotlib
- See the great user guide: https://matplotlib.org/

```python
# following line makes graphics work nicely within a Jupyter notebook:
%matplotlib inline 
# for plotting in Ipython, usually the pyplot module is loaded:
import matplotlib.pyplot as plt
# set default figure size:
plt.rcParams["figure.figsize"] = (14,7)
# Generate a sequence of numbers from -10 to 10 with 100 steps in between
x = np.linspace(-10, 10, 100) # also see np.arange
# Create a second array using the sine function
y = np.sin(x)
# printing out y will not be revealing:
print("y:\n{}".format(y))
# The plot function makes a line chart of one array against another
plt.plot(x, y, marker="x")
```

```python
# We can control how plots appear
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (10, 5),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
plt.rcParams.update(params)
fig, ax = plt.subplots()
plt.plot(x, y, marker="x")
ax.set(ylabel='Sine(x)',xlabel='x')
ax.set_xlim(-10,10)
fig.tight_layout()
plt.grid()
```

# Python control structures:
- if
- for loops
- while loops

```python
# Note the spaces in the following!!
if np.amax(y) > 0.5:
    y_small = 0.5 * y / np.amax(y)
else:
    y_small = y
plt.plot(x, y, marker="x")
plt.plot(x, y_small, marker="x")
plt.grid()
```

# For loops:

```python
expList = [1,3,4]
for exponent in expList:
    y_exp = np.power(y,exponent)
    plt.plot(x, y_exp, marker="x")
plt.legend(expList)
plt.grid()
```

# While-loop

- The example below can be solved using numpy without a loop.

```python
index = 0
location = x[index]
y_switch = np.zeros(y.shape)
while location < 0:
    y_switch[index] = y[index]
    index += 1
    location = x[index]
plt.plot(x, y_switch, marker="x")
plt.grid()
```
